### Finding Missing Values

##### Complete Data - Count


```python
df.createOrReplaceTempView('Spotify')
Count = spark.sql("Select count(*) from Spotify")
Count.show(1)
```

#### All Missing

```python
from pyspark.sql.functions import col, sum, when

missing_df = df.select([
    sum(when(col(c).isNull() | (col(c) == ""), 1).otherwise(0)).alias(c)
    for c in df.columns
])

missing_df.show()
```
