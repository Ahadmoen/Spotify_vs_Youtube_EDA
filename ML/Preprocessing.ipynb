### Data Loading

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("SparkSQLExample").getOrCreate()
```

```python
df = spark.read.csv('../Data/Spotify_Youtube.csv', header=True, inferSchema=True)
```

```python
df.show(2)
```

### Finding Missing Values

```python
df.createOrReplaceTempView('Spotify')
```

#### All Missing

```python
All = spark.sql("SELECT count(*) FROM Spotify")
All.show()
```

```python
result = spark.sql("""
SELECT count(*) from Spotify
where Artist is null""")
result.show(10)
```
